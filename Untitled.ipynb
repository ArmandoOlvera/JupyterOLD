```python
import numpy as np


class Perceptron(object):
    """Perceptron classifier. """
    

    def __init__(self, eta=0.01, n_iter=50, random_state=1):
        """
        Perceptron classifier.
        
        :param eta: (float) Learning rate (between 0.0 and 1.0)
        :param n_iter: (int) Passes over the training dataset
        :param random_state: (int) Random number generator seed for random weight initialization.
        
        """
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state
        self.x_pesos = [0,0,0]#para el AND
        self.x_pesos2 = [0,0,0]#PARA EL OR
        self.x3=np.array([[0,0],[0,0],[0,0],[0,0]])#Se define la estructura XOR esperando a ser llenada
        self.y3= np.array([0,0,0,0])#se gaurdan aqui los resultados que se obtendran
        self.w_ = None
        self.errors_ = None
        
    def fit(self, x, y):
        """
        
        Fit training data.
        
        :param x: {array-like, shape=[n_samples, n_features]} Training vectors, where n_samples is the number of
        samples and n_features is the number of features.
        :param y: {array-like, shape=n_samples} Target values.
        :rtype: Perceptron
        :return: self
        
        """
        ##Cracteristicas  w= n+1
        rgen = np.random.RandomState(self.random_state)
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=1 + x.shape[1])
        print(str(self.w_))
        self.errors_ = []
        
        for _ in range(self.n_iter):
            errors = 0
            for xi, target in zip(x, y):
                update = self.eta * (target - self.predict(xi))
                print("Update  "+str(update))
                self.w_[1:] += update * xi
                self.w_[0] += update
                print(str(self.w_))
                errors += int(update != 0.0)
            self.errors_.append(errors)
        return self
    
    def net_input(self, x):
        """
        
        Calculate net input
        
        :param x: {array-like, shape=[n_samples, n_features]} Training vectors, where n_samples is the number of
        samples and n_features is the number of features.
        :return:
        """
        
        return np.dot(x, self.w_[1:]) + self.w_[0]
    
    def predict(self, x):
        """
        Return class label after unit step
        
        :param x: {array-like, shape=[n_samples, n_features]} Training vectors, where n_samples is the number of
        samples and n_features is the number of features.
        :return:
        
        """
        return np.where(self.net_input(x) >= 0.0, 1, -1)
    
    def calcularGrafica(self, X,funcion):
        if funcion==1:
            z= (np.dot(X , self.x_pesos[1:]) + self.x_pesos[0])
        if funcion==2:
            z= (np.dot(X , self.x_pesos2[1:]) + self.x_pesos2[0])
        return np.where(z >= 0.0, 1, -1)
    
    def calcularXOR(self):
        for i in range(0,4):
            if self.x3[i,0]==self.x3[i,1]:
                self.y3[i]=-1
            else:
                self.y3[i]=1
            
                
    ##Aqui se crea la funcion que en base a una tabla de verdad y pesos se puede
    def entrenar(self, x,y,epocas,instancias,w0,w1,w2,aprendizaje,funcion):
        #valores iniciales
        errores=0
        on=1
        w = instancias+1
        z = 1#porque en la primera vuelta sale 1 y se tiene que entrenar
        posA=1 ## posicion de aprendizaje
        for e in range(0,epocas):
            
            print("epoca: "+str(e+1))
            for i in range(0,instancias):
                print("====================================================ITERACION===============================: "+str(1+i))
                print(str(w0)+"  "+str(w1)+"  "+str(w2))
                z=(1*w0)+(w1*x[i,0])+(w2*x[i,1])
                
                print("SUMATORIA Z=="+ str(z) )
                if z> 0:
                 z=1
                else:
                 z=-1
                print("Z ES:"+str(z))
                print("Y esperada: "+str(y[i]))
                ##print("y obtenida: "+str(z)+ " y esperada: "+str(y[i]))
                if z!= y[i]:
                    print("error")
                    errores=errores+1
                if funcion==1:
                    ##em caso de que se trate del AND se llenan los resultados de AND para la entrada de XOR
                    self.x3[i,0]=z
                    if self.x3[i,0]==-1:
                        self.x3[i,0]=0
                if funcion==2:
                    ##em caso de que se trate del OR se llenan los resultados de OR para la entrada de XOR
                    self.x3[i,1]=z
                    if self.x3[i,1]==-1:
                        self.x3[i,1]=0
                print(str(w0)+"  "+str(w1)+"  "+str(w2))
                if  i!=4:
                    w0=w0+(aprendizaje*(y[i]-z)*1)
                    w0=round(w0,2)
                if x[i,0]==1 and i!=4:
                    w1=w1+(aprendizaje*(y[i]-z)*x[i,0])
                if x[i,1]==1 and i!=4:
                    w2=w2+(aprendizaje*(y[i]-z)*x[i,1])
                print(str(w0)+"  "+str(w1)+"  "+str(w2))
                print(str(x[i,0])+" "+str(x[i,1]))
                
                
            print("Suma de errores por las 4 instancias: "+str(errores))
            if errores == 0:
                print("ya aprendi!!!")
                print("Con los pesos de: W0 = "+str(w0)+" W1= "+str(w1)+" w2= "+str(w2))
                if funcion==1:
                    self.x_pesos[0]=w0
                    self.x_pesos[1]=w1
                    self.x_pesos[2]=w2
                if funcion==2:
                    self.x_pesos2[0]=w0
                    self.x_pesos2[1]=w1
                    self.x_pesos2[2]=w2
                print("En la epoca "+str(e+1))
                break ##break porque no queremos seguir continuando lol
            else:
                print("me equivoque, no he aprendido!")
                errores=0
                
```

```python
def plot_decision_regions(X, y, classifier, resolution=0.02,funcion=1):

    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    if funcion!=3:
        Z = classifier.calcularGrafica(np.array([xx1.ravel(), xx2.ravel()]).T,funcion)
        Z = Z.reshape(xx1.shape)
    else:
        Z = classifier.calcularGrafica(np.array([xx1.ravel(), xx2.ravel()]).T,1)
        Z = Z.reshape(xx1.shape)
        plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
        Z = classifier.calcularGrafica(np.array([xx1.ravel(), xx2.ravel()]).T,2)
        Z = Z.reshape(xx1.shape)
    
    plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    # plot class samples
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
                    alpha=0.8, c=cmap(idx),
                    edgecolor='black',
                    marker=markers[idx], 
                    label=cl)
```
